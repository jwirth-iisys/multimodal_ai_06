```python
import soundfile as sf
from espnet2.bin.tts_inference import Text2Speech
from olaph import Olaph
import IPython

phonemizer = Olaph()

base_path = "/training-1/modelhub/vits/bernd-ungerer-22khz-updated"

config_path = f"{base_path}/config.yaml"
model = Text2Speech.from_pretrained(device="cuda",model_file=f"{base_path}/checkpoint.pth", train_config=config_path)

input_text = "Korruptionsermittlungen erschüttern die Kyjiwer Regierung. Im Mittelpunkt: Ein ehemaliger Geschäftspartner des Präsidenten."

input_phonemized = phonemizer.phonemize_text(input_text, lang="de")

print(input_phonemized)

speech = model(", "+input_phonemized)["wav"]
output_path = "output_vits.wav"
sf.write(output_path, speech.cpu().numpy(), model.fs, "PCM_16")

IPython.display.Audio(output_path)
```
